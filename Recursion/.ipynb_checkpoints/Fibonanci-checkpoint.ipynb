```python
import time
```

```python
def fib_rec(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return fib_rec(n-1) + fib_rec(n-2)
```

```python
def fib_rec_dynamic(n):
    if n <= 1:
        return (n,0)
    else:
        (a,b) = fib_rec_dynamic(n-1)
        return (a+b,a)
```

```python
start = time.time()
fib_rec_dynamic(2000)
end = time.time()
print(end - start)
```

```python
start = time.time()
fib_rec(100)
end = time.time()
print(end - start)
```

```python
cache = [None] * 12

def fib_dyn(n):
    if n == 0 or n == 1:
        return n
    if cache[n] != None:
        return cache[n]
    cache[n] = fib_dyn(n-1) + fib_dyn(n-2)
    return cache[n]
```

```python
fib_dyn(10)
```

```python
def good_fib(n):
    if n <= 1:
        return (n,0)
    else:
        (a,b) = good_fib(n-1)
        return (a+b,a)
```

```python
start = time.time()
good_fib(100)
end = time.time()
print(end - start)
```
